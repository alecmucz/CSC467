Quadratic Equation : $$ y = x^2 + 8x + 12 $$


```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(-1000, 1000, 1000)

# Quadratic equation: y = x^2 + 8x + 12
y = x**2 + 8*x + 12

# Plotting the graph
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^2 + 8x + 12', color='blue')
plt.title('Quadratic Equation Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
